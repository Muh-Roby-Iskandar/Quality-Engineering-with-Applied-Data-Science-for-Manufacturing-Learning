```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import BayesianGaussianMixture
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv(r"C:\Users\Robyi\Documents\Data Science Dataset\boston.csv")
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.head()
```

```python
df = df.select_dtypes(include=["number"]).dropna()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
bgmm = BayesianGaussianMixture(
    n_components=10,
    covariance_type="full",
    weight_concentration_prior_type="dirichlet_process",
    random_state=42
)
```

```python
bgmm.fit(X_scaled)
```

```python
cluster_labels = bgmm.predict(X_scaled)
```

```python
print(df["Cluster"].value_counts())
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=cluster_labels, palette="tab10")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Clustering dengan Bayesian Gaussian Mixture Model")
plt.legend(title="Cluster")
plt.show()
```
