```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv(r"C:\Users\Robyi\Documents\Data Science Dataset\wine.csv")
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df["Alcohol"]
y = df["Malic_Acid"]
z = df["Hue"]

ax.scatter(x, y, z, c=z, cmap='coolwarm', s=50)

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Hue")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(df["Alcohol"], df["Malic_Acid"], df["Hue"], marker='o')

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Hue")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(df["Alcohol"], df["Malic_Acid"])
Z = np.sin(X) + np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Function")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='black')

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Function")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df["Alcohol"][:10] 
y = df["Malic_Acid"][:10]
z = np.zeros(10)
dx = dy = np.ones(10)
dz = df["Hue"][:10] 

ax.bar3d(x, y, z, dx, dy, dz, shade=True)

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Hue")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df["Alcohol"][:10]
y = df["Malic_Acid"][:10]
z = df["Hue"][:10]
u = df["Ash"][:10] 
v = df["Magnesium"][:10] 
w = df["Hue"][:10] 
ax.quiver(x, y, z, u, v, w, length=0.5, normalize=True)

ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic Acid")
ax.set_zlabel("Hue")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.indices((3, 3, 3))

voxel_array = (x < 2) & (y < 2) & (z < 2)

ax.voxels(voxel_array, facecolors='red', edgecolors='black')

plt.show()
```
