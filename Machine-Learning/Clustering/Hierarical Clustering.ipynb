```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, silhouette_samples
from sklearn.manifold import TSNE
```

```python
df = df.drop(columns=['CHAS'])
```

```python
x = df
```

```python
linkage_methods = ['ward','complete', 'average', 'single']
distance_metrics = ['euclidean', 'manhattan']
n_clusters_range = range(2, 5)

best_score = -1
best_params = {}

for linkage in linkage_methods:
     for metric in distance_metrics:
          if linkage == 'ward' and metric != 'euclidean':
              continue
          for n_clusters in n_clusters_range:
            model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage, metric=metric)
            labels = model.fit_predict(x)
            score = silhouette_score(x, labels)
            if score > best_score:
                best_score = score
                best_params = {'linkage': linkage, 'metric': metric, 'n_clusters': n_clusters}

print("Best Parameters:", best_params)
print("Best Silhouette Score:", best_score)
```

```python
linked = linkage(x, method ='complete')
plt.figure(figsize=(20,16))
dendrogram(linked, orientation='top', distance_sort ='descending', show_leaf_counts=True)
plt.title('dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
silhouette_scores = []
cluster_range = range(2, 10)

for k in cluster_range:
    model = AgglomerativeClustering(n_clusters=k, metric='manhattan', linkage='complete')
    labels = model.fit_predict(x)
    score = silhouette_score(x, labels)
    silhouette_scores.append(score)

plt.plot(cluster_range, silhouette_scores, marker='o')
plt.xlabel("Number of Clusters (n_clusters)")
plt.ylabel("Silhouette Score")
plt.title("Optimal n_clusters dengan Silhouette Score")
plt.show()

optimal_clusters = cluster_range[np.argmax(silhouette_scores)]
print(f"Jumlah cluster optimal: {optimal_clusters}")
```
