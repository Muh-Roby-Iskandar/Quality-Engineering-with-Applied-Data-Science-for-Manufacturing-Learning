```python
import numpy as np
import random
from scipy.optimize import dual_annealing
```

```python
# ================== DATA ==================
# Matriks jarak antar kota
distance_matrix = np.array([
    [0, 12, 10, 19, 8, 15],  # A
    [12, 0, 3, 7, 2, 6],      # B
    [10, 3, 0, 4, 5, 9],      # C
    [19, 7, 4, 0, 10, 12],    # D
    [8, 2, 5, 10, 0, 11],     # E
    [15, 6, 9, 12, 11, 0]     # F
])
```

```python
num_cities = distance_matrix.shape[0]  # Jumlah kota
city_names = ["A", "B", "C", "D", "E", "F"]
```

```python
# ================== FUNGSI FITNESS ==================
def total_distance(route):
    """Menghitung total jarak dari rute tertentu."""
    route = np.round(route).astype(int)  # Pastikan index adalah integer
    total_dist = 0
    for i in range(len(route) - 1):
        total_dist += distance_matrix[route[i], route[i + 1]]
    total_dist += distance_matrix[route[-1], route[0]]  # Kembali ke kota asal
    return total_dist
```

```python
# ================== KONFIGURASI SIMULATED ANNEALING ==================
# Kita optimasi urutan kota yang dikunjungi
bounds = [(0, num_cities - 1)] * num_cities  # Setiap kota dalam batas indeks [0, num_cities-1]

result = dual_annealing(total_distance, bounds, maxiter=5000)
```

```python
# ================== OUTPUT HASIL ==================
best_route = np.round(result.x).astype(int)  # Konversi hasil ke indeks kota
best_route_names = [city_names[i] for i in best_route]
best_distance = result.fun
```

```python
print(f"Rute terbaik: {best_route_names}")
print(f"Total jarak terbaik: {best_distance} km")
```
