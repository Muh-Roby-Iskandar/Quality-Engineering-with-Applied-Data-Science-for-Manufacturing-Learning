```python
import itertools

def sliding_window(seq, n=3):
    iters = itertools.tee(seq, n)
    for i, it in enumerate(iters):
        for _ in range(i):
            next(it, None)
    return zip(*iters)

data = [10, 20, 30, 40, 50]

for window in sliding_window(data, 3):
    print(window)

# contoh 3
```

```python
from itertools import groupby

data = [
    ("A", 10), ("A", 15),
    ("B", 5), ("B", 7), ("B", 20),
]

# Data harus disortir berdasarkan key
data_sorted = sorted(data, key=lambda x: x[0])

for key, group in groupby(data_sorted, key=lambda x: x[0]):
    values = [v for _, v in group]
    print(key, sum(values))
```
