```python
import numpy as np
```

1. Single Sampling Plan

```python
# Parameter
sample_size = 50
acceptance_number = 3

# Simulasi jumlah cacat dalam sampel
np.random.seed(42)
defective_units = np.random.randint(0, 6, sample_size)  # Random cacat (0-5 per unit)
total_defects = sum(defective_units > 0)  # Hitung unit cacat

# Keputusan Acceptance Sampling
if total_defects <= acceptance_number:
    print(f"Batch DITERIMA (Defects: {total_defects})")
else:
    print(f"Batch DITOLAK (Defects: {total_defects})")
```

2. Double Sampling Plan

```python
# Parameter Double Sampling
sample1_size, c1, r1 = 30, 2, 5
sample2_size, c2, r2 = 20, 5, 6

# Simulasi jumlah cacat
np.random.seed(42)
defects_sample1 = np.random.randint(0, 6, sample1_size).sum()
defects_sample2 = np.random.randint(0, 6, sample2_size).sum()

# Keputusan
if defects_sample1 <= c1:
    print(f"Batch DITERIMA setelah Sampel 1 (Defects: {defects_sample1})")
elif defects_sample1 >= r1:
    print(f"Batch DITOLAK setelah Sampel 1 (Defects: {defects_sample1})")
else:
    total_defects = defects_sample1 + defects_sample2
    if total_defects <= c2:
        print(f"Batch DITERIMA setelah Sampel 2 (Total Defects: {total_defects})")
    else:
        print(f"Batch DITOLAK setelah Sampel 2 (Total Defects: {total_defects})")
```

3. Multiple Sampling Plan

```python
# Parameter Multiple Sampling
stages = 3
sample_sizes = [20, 20, 20]  # Total 60 unit diperiksa dalam 3 tahap
acceptance_limits = [1, 3, 5]
rejection_limits = [4, 6, 8]

np.random.seed(42)
total_defects = 0

for i in range(stages):
    defects = np.random.randint(0, 6, sample_sizes[i]).sum()
    total_defects += defects
    print(f"Stage {i+1}: Defects = {defects}, Total Defects = {total_defects}")

    if total_defects <= acceptance_limits[i]:
        print(f"Batch DITERIMA setelah Stage {i+1}")
        break
    elif total_defects >= rejection_limits[i]:
        print(f"Batch DITOLAK setelah Stage {i+1}")
        break
```

4. Sequential Sampling Plan

```python
# Parameter Sequential Sampling
max_units = 50
acceptance_limit = 2
rejection_limit = 5

np.random.seed(42)
total_defects = 0

for i in range(1, max_units + 1):
    defect = np.random.randint(0, 2)  # 0 = baik, 1 = cacat
    total_defects += defect
    print(f"Unit {i}: Defects = {total_defects}")

    if total_defects >= rejection_limit:
        print(f"Batch DITOLAK setelah {i} unit diperiksa")
        break
    elif total_defects <= acceptance_limit and i >= 10:
        print(f"Batch DITERIMA setelah {i} unit diperiksa")
        break
```
