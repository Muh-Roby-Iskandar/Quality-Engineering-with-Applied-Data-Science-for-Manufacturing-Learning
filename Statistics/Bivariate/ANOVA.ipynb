```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Dokumen\Python\Data Engineering 1\Wine.csv")
```

```python
from scipy.stats import f_oneway
```

```python
df.head()
```

```python
group = [df[df['Ash'] == q]
         ['Alcohol'] for q in 
         df['Ash'].unique()]
```

```python
group
```

```python
f_stat, p_value = f_oneway(*group)
```

```python
print(f_stat)
print(p_value)
```

```python
if p_value < 0.05:
    print("Signifikan")
else:
    print("Tidak Signifikan")
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('Customer_Segment ~ C(Ash) + C(Magnesium) + C(Ash):C(Magnesium)', data = df).fit()
```

```python
anova_table = sm.stats.anova_lm(model, typ=2)
```
