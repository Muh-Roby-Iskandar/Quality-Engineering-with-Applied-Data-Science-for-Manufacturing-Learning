```python
from sklearn.decomposition import FactorAnalysis
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = load_iris()
```

```python
X = data.data 
```

```python
n_factors = 2
fa = FactorAnalysis(n_components=n_factors)
```

```python
X_fa = fa.fit_transform(X)
```

```python
df_fa = pd.DataFrame(X_fa, columns=[f'Factor{i+1}' for i in range(n_factors)])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_fa['Factor1'], df_fa['Factor2'], c='blue', edgecolor='k', alpha=0.5)
plt.title('Factor Analysis (2 Factors)')
plt.xlabel('Factor 1')
plt.ylabel('Factor 2')
plt.grid(True)
plt.show()
```

```python
print("Explained Variance Ratio:", fa.components_)
```
