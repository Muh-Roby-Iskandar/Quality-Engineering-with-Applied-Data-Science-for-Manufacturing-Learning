```python
import numpy as np
import pandas as pd
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.array([
    [0, 10, 20, 30],
    [10, 0, 15, 25],
    [20, 15, 0, 35],
    [30, 25, 35, 0]
])
```

```python
labels = ['Objek_1', 'Objek_2', 'Objek_3', 'Objek_4']
```

```python
df_distance = pd.DataFrame(data, index=labels, columns=labels)
print("Matriks Jarak:")
print(df_distance)
```

```python
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
mds_coords = mds.fit_transform(data)
```

```python
df_mds_coords = pd.DataFrame(mds_coords, columns=['Dimensi_1', 'Dimensi_2'], index=labels)
print("\nKoordinat Hasil MDS:")
print(df_mds_coords)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Dimensi_1', y='Dimensi_2', data=df_mds_coords, s=100, color='blue')
for label in df_mds_coords.index:
    plt.text(df_mds_coords.loc[label, 'Dimensi_1'], 
             df_mds_coords.loc[label, 'Dimensi_2'], 
             label, fontsize=12, color='black')
plt.title('Visualisasi Multi-Dimensional Scaling (MDS)', fontsize=14)
plt.xlabel('Dimensi 1', fontsize=12)
plt.ylabel('Dimensi 2', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.show()
```
