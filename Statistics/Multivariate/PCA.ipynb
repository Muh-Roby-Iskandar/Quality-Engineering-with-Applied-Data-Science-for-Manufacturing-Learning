```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = load_iris()
```

```python
X = data.data 
y = data.target  
```

```python
feature_names = data.feature_names 
class_names = data.target_names 
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
```

```python
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca['Target'] = y
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, color='skyblue')
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', color='orange')
plt.title('Scree Plot')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
loadings = pca.components_.T 
plt.figure(figsize=(10, 6))
for i, feature in enumerate(feature_names):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='r', head_width=0.02)
    plt.text(loadings[i, 0] * 1.1, loadings[i, 1] * 1.1, feature, color='b')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Loading Plot (PCA)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df_pca['PCA1'], df_pca['PCA2'], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA Projection (2D)')
plt.colorbar(scatter, label='Class')
plt.grid()
plt.show()
```
