```python
from scipy.optimize import linprog
```

```python
# Parameter
costs = [10, 12, 15]  # Biaya produksi per bulan
demand = [100, 150, 200]  # Permintaan bulanan
max_production = 180  # Kapasitas produksi maksimum
```

```python
# Kendala produksi
A_eq = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
b_eq = demand
```

```python
# Batasan produksi
bounds = [(0, max_production)] * 3
```

```python
# Optimasi LP
res = linprog(costs, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method="highs")
```

```python
print("Hasil Aggregate Planning:")
print(f"Produksi bulan 1: {res.x[0]:.0f} unit")
print(f"Produksi bulan 2: {res.x[1]:.0f} unit")
print(f"Produksi bulan 3: {res.x[2]:.0f} unit")
print(f"Total Biaya: {res.fun:.2f}")
```
