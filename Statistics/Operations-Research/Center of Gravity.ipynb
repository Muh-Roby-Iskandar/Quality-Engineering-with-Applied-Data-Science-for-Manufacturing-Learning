```python
import numpy as np
```

```python
# Data lokasi (x, y) dan volume pengiriman
locations = np.array([
    [2, 3, 500],  # (x, y, demand)
    [5, 7, 700],
    [8, 2, 300],
    [6, 6, 600]
])
```

```python
# Hitung koordinat pusat gravitasi
x_cog = np.sum(locations[:, 0] * locations[:, 2]) / np.sum(locations[:, 2])
y_cog = np.sum(locations[:, 1] * locations[:, 2]) / np.sum(locations[:, 2])
```

```python
print(f"Center of Gravity Location: ({x_cog:.2f}, {y_cog:.2f})")
```
