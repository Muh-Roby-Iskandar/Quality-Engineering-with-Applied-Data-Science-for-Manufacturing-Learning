```python
# Fungsi untuk menghitung total biaya material handling
def calculate_total_cost(layout, flow_matrix):
    total_cost = 0
    for i in range(len(layout)):
        for j in range(len(layout)):
            total_cost += flow_matrix[i][j] * abs(layout[i] - layout[j])
    return total_cost
```

```python
# Contoh layout awal
layout = [0, 1, 2]
print("Total Cost:", calculate_total_cost(layout, flow_matrix))
```

```python
# Visualisasi layout
plt.figure(figsize=(6, 4))
plt.bar(range(len(layout)), layout, color='skyblue')
plt.title("Facility Layout")
plt.xlabel("Facility")
plt.ylabel("Position")
plt.show()
```

```python
# Fungsi untuk menghitung total biaya material handling
def calculate_total_cost(layout, flow_matrix, distance_matrix):
    total_cost = 0
    for i in range(len(layout)):
        for j in range(len(layout)):
            total_cost += flow_matrix[i][j] * distance_matrix[layout[i]][layout[j]]
    return total_cost
```

```python
# Contoh data
flow_matrix = np.array([
    [0, 10, 5],
    [10, 0, 8],
    [5, 8, 0]
])
```

```python
distance_matrix = np.array([
    [0, 2, 3],
    [2, 0, 1],
    [3, 1, 0]
])
```

```python
# Contoh layout awal
layout = [0, 1, 2]
print("Total Cost:", calculate_total_cost(layout, flow_matrix, distance_matrix))
```

```python
# Pertukaran posisi fasilitas (contoh heuristik sederhana)
def swap_facilities(layout, i, j):
    layout[i], layout[j] = layout[j], layout[i]
    return layout
```

```python
# Contoh pertukaran
new_layout = swap_facilities(layout, 0, 1)
print("New Layout:", new_layout)
print("New Total Cost:", calculate_total_cost(new_layout, flow_matrix, distance_matrix))
```
