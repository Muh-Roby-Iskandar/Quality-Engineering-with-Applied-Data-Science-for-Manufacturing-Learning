1. From-To Chart

```python
import pandas as pd
import numpy as np
```

```python
# Data aliran antar departemen (From-To Chart)
departments = ["A", "B", "C", "D"]
flow_matrix = np.array([
    [0, 30, 10, 20],  # Dari A ke lainnya
    [30, 0, 50, 10],  # Dari B ke lainnya
    [10, 50, 0, 40],  # Dari C ke lainnya
    [20, 10, 40, 0]   # Dari D ke lainnya
])
```

```python
# Buat DataFrame
from_to_chart = pd.DataFrame(flow_matrix, index=departments, columns=departments)
print("From-To Chart:")
print(from_to_chart)
```

2. Analisis Momen Trial (Menentukan Tata Letak Optimal)

```python
# Koordinat awal departemen (misal grid 2D)
positions = {
    "A": (0, 0),
    "B": (2, 0),
    "C": (0, 2),
    "D": (2, 2)
}
```

```python
# Hitung momen untuk tiap departemen
moments = {}
for dept1 in departments:
    mx, my, total_flow = 0, 0, 0
    x1, y1 = positions[dept1]
    for dept2 in departments:
        if dept1 != dept2:
            x2, y2 = positions[dept2]
            flow = flow_matrix[departments.index(dept1), departments.index(dept2)]
            mx += flow * x2
            my += flow * y2
            total_flow += flow
    moments[dept1] = (mx / total_flow if total_flow else x1, my / total_flow if total_flow else y1)
```

```python
# Hasil tata letak baru berdasarkan analisis momen
print("\nAnalisis Momen Trial (Koordinat Baru):")
for dept, pos in moments.items():
    print(f"{dept}: {pos}")
```
