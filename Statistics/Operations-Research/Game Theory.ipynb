```python
import numpy as np
from scipy.optimize import linprog
```

```python
# Matriks pembayaran untuk pemain A
payoff_A = np.array([[3, -1], [0, 2]])  # Baris: strategi A, Kolom: strategi B
```

```python
# Maksimalkan keuntungan pemain A
c = [-1, -1]  # Maksimalkan -> minimalkan negatif
A_eq = [[1, 1]]  # Probabilitas strategi A1 + A2 = 1
b_eq = [1]
bounds = [(0, 1)] * 2
```

```python
# Optimasi strategi campuran
res = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method="highs")
```

```python
print("Strategi Optimal Pemain A:")
print(f"Peluang memilih strategi A1: {res.x[0]:.2f}")
print(f"Peluang memilih strategi A2: {res.x[1]:.2f}")
```
