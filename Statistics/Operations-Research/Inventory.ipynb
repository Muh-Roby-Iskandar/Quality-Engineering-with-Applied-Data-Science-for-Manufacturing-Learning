1. EOQ (Economic Order Quantity)

```python
import numpy as np
```

```python
# Parameter EOQ
D = 1000  # Permintaan tahunan
S = 50    # Biaya pemesanan per order
H = 2     # Biaya penyimpanan per unit per tahun
```

```python
# Rumus EOQ
EOQ = np.sqrt((2 * D * S) / H)
TBO = EOQ / D  # Time Between Orders
```

```python
print(f"EOQ: {EOQ:.2f} unit")
print(f"TBO: {TBO:.2f} tahun")
```

2. POQ (Production Order Quantity)

```python
# Parameter POQ
p = 2000  # Produksi per tahun
c = 100   # Biaya setup produksi
```

```python
# Rumus POQ
POQ = np.sqrt((2 * D * c) / (H * (1 - D/p)))
TBO_POQ = POQ / D  # Time Between Orders
```

```python
print(f"POQ: {POQ:.2f} unit")
print(f"TBO_POQ: {TBO_POQ:.2f} tahun")
```
