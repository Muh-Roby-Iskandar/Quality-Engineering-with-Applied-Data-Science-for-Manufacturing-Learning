```python
import numpy as np
import pandas as pd
from scipy.optimize import linprog
```

MAKSIMASI

```python
# Koefisien fungsi objektif (Maximize Z = 3x1 + 5x2)
c = [-3, -5]  # Linprog meminimalkan, jadi negatif

# Koefisien kendala
A = [
    [2, 3],   # 2x1 + 3x2 <= 8
    [1, 2]    # x1 + 2x2 <= 4
]
b = [8, 4]  # Sisi kanan kendala
```

```python
# Batasan variabel (x1, x2 ≥ 0)
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]
```

```python
# Menjalankan Simplex
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method="highs")
```

```python
# Menampilkan hasil
print("Nilai maksimum Z:", -result.fun)
print("Nilai x1:", result.x[0])
print("Nilai x2:", result.x[1])
```

MINIMASI

```python
# Koefisien fungsi objektif (Minimize Z = 4x1 + 6x2)
c = [4, 6]  # Tidak perlu diubah tanda karena minimisasi

# Koefisien kendala (ubah menjadi <= dengan mengalikan -1)
A = [
    [-3, -2],  # 3x1 + 2x2 >= 12  →  -3x1 - 2x2 <= -12
    [-1, -4]   # x1 + 4x2 >= 8   →  -x1 - 4x2 <= -8
]
b = [-12, -8]  # Sesuai perubahan tanda
```

```python
# Batasan variabel (x1, x2 ≥ 0)
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]
```

```python
# Menjalankan linprog
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method="highs")
```

```python
# Menampilkan hasil
print("Nilai minimum Z:", result.fun)
print("Nilai x1:", result.x[0])
print("Nilai x2:", result.x[1])
```
