```python
import numpy as np
```

```python
# Matriks transisi
P = np.array([[0.7, 0.3], 
              [0.4, 0.6]])
```

```python
# Distribusi awal
state_0 = np.array([1, 0])  # Awalnya di state 1
```

```python
# Prediksi setelah 3 periode
state_3 = np.linalg.matrix_power(P, 3).dot(state_0)
```

```python
print("Distribusi Probabilitas setelah 3 periode:", state_3)
```
