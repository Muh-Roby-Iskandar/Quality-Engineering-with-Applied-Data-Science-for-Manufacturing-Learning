```python
import pandas as pd
```

```python
# Struktur produk (BOM)
BOM = {
    "Produk_A": {"Komponen_X": 2, "Komponen_Y": 1},
    "Komponen_X": {"Bahan_1": 3, "Bahan_2": 2},
    "Komponen_Y": {"Bahan_2": 4, "Bahan_3": 1},
}
```

```python
# Permintaan produk akhir
demand = {"Produk_A": 50}
```

```python
# Hitung kebutuhan material
def compute_MRP(demand, BOM):
    requirements = {}
    to_process = list(demand.items())

    while to_process:
        item, qty = to_process.pop(0)
        if item in requirements:
            requirements[item] += qty
        else:
            requirements[item] = qty
        
        if item in BOM:
            for component, amount in BOM[item].items():
                to_process.append((component, qty * amount))

    return requirements
```

```python
mrp_result = compute_MRP(demand, BOM)
print("\nMaterial Requirements Planning (MRP):")
print(pd.DataFrame(mrp_result.items(), columns=["Item", "Total Kebutuhan"]))
```
