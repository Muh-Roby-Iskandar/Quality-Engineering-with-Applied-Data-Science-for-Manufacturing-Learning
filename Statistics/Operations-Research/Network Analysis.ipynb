1. Shortest Path (Dijkstra Algorithm)

```python
import networkx as nx
```

```python
# Buat graf dengan bobot
G = nx.DiGraph()
edges = [(1, 2, 4), (1, 3, 2), (2, 3, 5), (2, 4, 10), (3, 4, 3)]
G.add_weighted_edges_from(edges)
```

```python
# Jalankan Dijkstra dari node 1 ke 4
shortest_path = nx.shortest_path(G, source=1, target=4, weight="weight")
shortest_distance = nx.shortest_path_length(G, source=1, target=4, weight="weight")
```

```python
print("Shortest Path (Dijkstra):", shortest_path)
print("Shortest Distance:", shortest_distance)
```

2. Maximum Flow (Ford-Fulkerson)

```python
# Buat graf untuk max flow
G = nx.DiGraph()
edges = [(1, 2, 16), (1, 3, 13), (2, 3, 10), (2, 4, 12), (3, 2, 4), (3, 5, 14),
         (4, 3, 9), (4, 6, 20), (5, 4, 7), (5, 6, 4)]
G.add_weighted_edges_from(edges, weight="capacity")
```

```python
# Maksimum aliran dari 1 ke 6
flow_value, flow_dict = nx.maximum_flow(G, _s=1, _t=6)
print("\nMaximum Flow:", flow_value)
print("Flow Path:", flow_dict)
```

3. Minimum Spanning Tree (Prim's Algorithm)

```python
# Buat graf untuk MST
G = nx.Graph()
edges = [(1, 2, 4), (1, 3, 2), (2, 3, 5), (2, 4, 10), (3, 4, 3)]
G.add_weighted_edges_from(edges)
```

```python
# Minimum Spanning Tree dengan Prim
mst = nx.minimum_spanning_tree(G, algorithm="prim")
print("\nMinimum Spanning Tree (Prim):", mst.edges(data=True))
```

4. Dynamic Programming (Knapsack Problem)

```python
def knapsack(weights, values, W):
    n = len(values)
    dp = [[0] * (W + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(W + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][W]
```

```python
weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
W = 5
```

```python
print("\nMaximum Value (Knapsack):", knapsack(weights, values, W))
```
