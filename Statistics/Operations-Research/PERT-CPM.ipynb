```python
import networkx as nx
```

```python
# Buat graf aktivitas proyek
G = nx.DiGraph()
G.add_weighted_edges_from([
    ("Start", "A", 3), ("Start", "B", 2), ("A", "C", 4), ("B", "C", 6),
    ("C", "End", 5)
])
```

```python
# Critical Path
length, path = nx.single_source_dijkstra(G, "Start", "End", weight="weight")
print("\nCritical Path:", path)
print("Project Duration:", length)
```

```python
# Crashing: Misalnya kita percepat aktivitas B dari 2 menjadi 1
G["Start"]["B"]["weight"] = 1
length_crash, path_crash = nx.single_source_dijkstra(G, "Start", "End", weight="weight")
```

```python
print("\nAfter Crashing:")
print("New Critical Path:", path_crash)
print("New Project Duration:", length_crash)
```
