```python
import pandas as pd
import numpy as np
import pm4py
```

2. Membuat Data Event Log dengan Pandas

```python
np.random.seed(42)
```

```python
data = {
    "case_id": np.repeat(range(1, 6), 4),
    "activity": ["Start", "Check", "Approve/Reject", "End"] * 5,
    "timestamp": pd.date_range(start="2024-03-01", periods=20, freq="H"),
    "resource": np.random.choice(["Admin A", "Admin B", "Manager A", "Manager B"], size=20)
}
```

```python
event_log = pd.DataFrame(data)
```

```python
event_log = event_log.sort_values(by=["case_id", "timestamp"])
```

```python
event_log_pm4py = pm4py.format_dataframe(event_log, case_id="case_id", activity_key="activity", timestamp_key="timestamp")
```

```python
print(event_log_pm4py)
```

a) Directly-Follows Graph (DFG)

```python
dfg, start_activities, end_activities = pm4py.discover_dfg(event_log_pm4py)

pm4py.view_dfg(dfg, start_activities, end_activities)
```

b) Process Tree (Menggunakan Inductive Miner)

```python
process_tree = pm4py.discover_process_tree_inductive(event_log_pm4py)

pm4py.view_process_tree(process_tree)
```

c) Petri Net (Menggunakan Inductive Miner)

```python
net, im, fm = pm4py.discover_petri_net_inductive(event_log_pm4py)

pm4py.view_petri_net(net, im, fm)
```

d) Heuristic Net (Menggunakan Heuristic Miner)

```python
heuristic_net = pm4py.discover_heuristics_net(event_log_pm4py)

pm4py.view_heuristics_net(heuristic_net)
```

e) BPMN (Business Process Model and Notation)

```python
bpmn_model = pm4py.discover_bpmn_inductive(event_log_pm4py)

pm4py.view_bpmn(bpmn_model)
```
