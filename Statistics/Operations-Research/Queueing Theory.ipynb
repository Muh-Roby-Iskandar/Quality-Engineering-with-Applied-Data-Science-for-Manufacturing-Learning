```python
# Parameter antrian M/M/1
λ = 5  # Rata-rata kedatangan per jam
μ = 8  # Rata-rata layanan per jam
```

```python
# Rumus M/M/1
ρ = λ / μ  # Tingkat utilisasi
Lq = (λ**2) / (μ * (μ - λ))  # Rata-rata jumlah pelanggan dalam antrian
Wq = Lq / λ  # Rata-rata waktu tunggu dalam antrian
L = Lq + (λ / μ)  # Rata-rata jumlah pelanggan dalam sistem
W = Wq + (1 / μ)  # Rata-rata waktu dalam sistem
```

```python
print("\nHasil Analisis Antrian (M/M/1):")
print(f"Utilisasi sistem: {ρ:.2f}")
print(f"Jumlah rata-rata pelanggan dalam antrian: {Lq:.2f}")
print(f"Waktu rata-rata dalam antrian: {Wq:.2f} jam")
print(f"Jumlah rata-rata pelanggan dalam sistem: {L:.2f}")
print(f"Waktu rata-rata dalam sistem: {W:.2f} jam")
```
