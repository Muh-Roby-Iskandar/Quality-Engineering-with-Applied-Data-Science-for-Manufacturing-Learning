1. Simulasi Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Jumlah titik yang akan digunakan dalam simulasi
num_samples = 10000
```

```python
# Menghasilkan titik acak dalam rentang [-1, 1] untuk x dan y
x = np.random.uniform(-1, 1, num_samples)
y = np.random.uniform(-1, 1, num_samples)
```

```python
# Menghitung jarak titik dari pusat (0,0)
distance = np.sqrt(x*2 + y*2)
```

```python
# Menentukan titik mana yang berada dalam lingkaran satuan
inside_circle = distance <= 1
```

```python
# Estimasi nilai Ï€
pi_estimate = (np.sum(inside_circle) / num_samples) * 4
```

```python
# Visualisasi hasil simulasi
plt.figure(figsize=(6,6))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1, label="Inside Circle")
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1, label="Outside Circle")
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Monte Carlo Simulation for Pi Estimation\nEstimated Ï€ â‰ˆ {pi_estimate:.5f}")
plt.legend()
plt.axis("equal")
plt.show()
```

2. Diagram Pareto

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data masalah dan jumlah kejadian
problems = ["Kualitas Buruk", "Keterlambatan", "Kesalahan Produksi", "Bahan Rusak", "Kurang Efisien"]
counts = [50, 30, 20, 15, 10]
```

```python

# Mengurutkan data dari terbesar ke terkecil
sorted_indices = np.argsort(counts)[::-1]
sorted_problems = np.array(problems)[sorted_indices]
sorted_counts = np.array(counts)[sorted_indices]
```

```python
# Menghitung persentase kumulatif
cumulative_percentage = np.cumsum(sorted_counts) / np.sum(sorted_counts) * 100
```

```python
# Membuat diagram Pareto
fig, ax1 = plt.subplots(figsize=(8,5))

# Diagram batang (bar chart)
ax1.bar(sorted_problems, sorted_counts, color="blue", alpha=0.6, label="Frekuensi")
ax1.set_xlabel("Penyebab Masalah")
ax1.set_ylabel("Jumlah Kejadian", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# Diagram garis (cumulative percentage)
ax2 = ax1.twinx()
ax2.plot(sorted_problems, cumulative_percentage, color="red", marker="o", linestyle="-", label="Persentase Kumulatif")
ax2.set_ylabel("Persentase Kumulatif (%)", color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_ylim(0, 110)

# Menambahkan garis horizontal pada 80% (prinsip Pareto 80/20)
ax2.axhline(80, color='gray', linestyle='dashed', alpha=0.5)
plt.title("Diagram Pareto")
plt.show()
```
